```python
import pandas as pd
df=pd.read_csv('covid-19.zip')
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
total_states = np.arange(len(df['State/UTs'])) 
```

## total positive cases based on states in india

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(9, 6), dpi=80, facecolor='w', edgecolor='k')
plt.barh(total_states,df['Active'], align='center', alpha=0.5,  
                 color=(1,0,0),  
                 edgecolor=(0.5,0.2,0.8))
    
plt.yticks(total_states, df['State/UTs'])  
plt.xlim(1,max(df['Active'])+100) 
plt.xlabel('Active Number of Cases')  
plt.title('Corona Virus Cases')  
plt.show()
```

## dischared number of cases based on states

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(9, 6), dpi=80, facecolor='w', edgecolor='k')
plt.barh(total_states,df['Discharged'], align='center', alpha=0.5,  
                 color=(1,0.5,0),  
                 edgecolor=(0.5,0.4,0.8)  )
    
plt.yticks(total_states, df['State/UTs'])  
plt.xlim(1,max(df['Discharged'])+10) 
plt.xlabel('Discharged Number of Cases')  
plt.title('Corona Virus Cases')  
plt.show() 
```
